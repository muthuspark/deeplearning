```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.random.uniform(low=1, high=20, size=(50,))
y = 2*x + 1

plt.plot(x, y)
plt.show()
```

```python
# Sequntial - allows you to create models layer-by-layer for most problems
# Dense - a layer of connected neurons
# units = 1, since there is only one neuron here
# input_shape = [1], since the shape of the array is a 1D array
model = Sequential()
model.add(Dense(5, input_shape=(1,), activation='tanh'))
model.add(Dense(1, activation='linear'))
```

```python
# We use sgd optimizer which looks at mean squared error to improve the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=5000, verbose=0)
```

```python
# Evaluate the model
scores = model.evaluate(x, y, verbose=0)
print('%s: %.2f%%' % (model.metrics_names[1], scores[1] * 100))

# Make predictions
y_pred = model.predict(x)

new_x = np.arange(0, np.pi * 2, 0.01)
y_pred = model.predict(new_x)
plt.plot(new_x,y_pred)
plt.plot(x,y)
plt.show()
```
